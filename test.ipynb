```python
# 图片压缩测试
from PIL import Image
file_path = 'testImage/output.png'
with Image.open('testImage/c2b094ad-dc1c-494d-83e2-8f483d81728a.png') as img:
    # 计算等比例缩放后的尺寸
    ratio = 500/max(img.size)
    # ratio = 1
    new_size = (int(img.width * ratio), int(img.height * ratio))
    img = img.resize(new_size, Image.LANCZOS)
    img.save(file_path, quality=80, optimize=False)
```
